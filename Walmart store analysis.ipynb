# converting date columns

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data.info()
```

```python
data['Time'] = pd.to_datetime(data['Time'])
```

```python
data.info()
```

```python
data.head()
```

```python
data['Month'] = data['Date'].dt.month
data['Year'] = data['Date'].dt.year
data['Day of Week'] = data['Date'].dt.dayofweek
```

```python
data.head()
```

# Create time_of_day column

```python
def get_time_of_day(t):
    hour = pd.to_datetime(t, format="%H:%M:%S").hour
    if hour < 12:
        return "Morning"
    elif 12 <= hour < 16:
        return "Afternoon"
    else:
        return "Evening"

data['time_of_day'] = data['Time'].apply(get_time_of_day)
```

# Create day_name column

```python
data['day_name'] = data['Date'].dt.day_name()
```

# Create month_name column

```python
data['month_name'] = data['Date'].dt.month_name().str[:3]
```

```python
data[['Date', 'Time', 'time_of_day', 'day_name', 'month_name']].head()
```

```python
cities = data['City'].nunique()
cities
```

```python
branch_city = data[['Branch', 'City']].drop_duplicates()
branch_city
```

# Product Analysis

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
```

```python
data['Product line'].nunique()
```

```python
payment_counts = data['Payment'].value_counts()

plt.figure(figsize=(6,6))
payment_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title("Most Common Payment Method")
plt.ylabel("")
plt.show()

payment_counts
```

```python
product_sales = data.groupby('Product line')['Quantity'].sum().sort_values()

plt.figure(figsize=(8,5))
product_sales.plot(kind='barh')
plt.title("Best Selling Product Line (by Quantity)")
plt.xlabel("Total Quantity Sold")
plt.show()

product_sales
```

```python
monthly_revenue = data.groupby('month_name')['Total'].sum()

plt.figure(figsize=(8,5))
monthly_revenue.plot(kind='bar', edgecolor='black')
plt.title("Total Revenue by Month")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()

monthly_revenue
```

```python
month_cogs = data.groupby('month_name')['cogs'].sum().sort_values(ascending=False)
month_cogs
```

```python
pl_revenue = data.groupby('Product line')['Total'].sum().sort_values()

plt.figure(figsize=(8,5))
pl_revenue.plot(kind='barh', edgecolor='black')
plt.title("Revenue by Product Line")
plt.xlabel("Revenue")
plt.show()

pl_revenue
```

```python
city_rev = data.groupby('City')['Total'].sum().sort_values()

plt.figure(figsize=(7,4))
city_rev.plot(kind='bar')
plt.title("Revenue by City")
plt.ylabel("Revenue")
plt.show()

city_rev
```

```python
data.groupby('Product line')['Tax 5%'].sum().sort_values(ascending=False)
```

```python
avg_sales = data['Total'].mean()

data['product_category'] = data['Total'].apply(lambda x: 'Good' if x >= avg_sales else 'Bad')
data[['Product line', 'Total', 'product_category']].head()
```

```python
branch_qty = data.groupby('Branch')['Quantity'].sum()
avg_qty = branch_qty.mean()

branch_qty.plot(kind='bar', figsize=(7,4))
plt.axhline(avg_qty, color='red', linestyle='--', label='Average Sold')
plt.title("Branch Product Sales")
plt.ylabel("Quantity Sold")
plt.legend()
plt.show()

branch_qty[branch_qty > avg_qty]
```

```python
gender_pl = data.groupby(['Gender', 'Product line'])['Quantity'].sum()

gender_pl.unstack().plot(kind='bar', figsize=(10,6))
plt.title("Most Common Product Line by Gender")
plt.ylabel("Quantity Sold")
plt.show()

gender_pl
```

```python
avg_rating = data.groupby('Product line')['Rating'].mean().sort_values()

plt.figure(figsize=(8,5))
avg_rating.plot(kind='barh')
plt.title("Average Rating per Product Line")
plt.xlabel("Rating")
plt.show()

avg_rating
```

# Sales Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt

sales_time_day = data.pivot_table(values='Total',
                                index='day_name',
                                columns='time_of_day',
                                aggfunc='count')

plt.figure(figsize=(8,6))
sns.heatmap(sales_time_day, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Sales Count by Time of Day and Weekday")
plt.show()

sales_time_day
```

```python
rev_by_customer_type = data.groupby('Customer type')['Total'].sum()

plt.figure(figsize=(6,6))
rev_by_customer_type.plot(kind='pie', autopct='%1.1f%%')
plt.title("Revenue Contribution by Customer Type")
plt.ylabel("")
plt.show()

rev_by_customer_type
```

```python
vat_city = data.groupby('City')['Tax 5%'].sum().sort_values()

plt.figure(figsize=(7,4))
vat_city.plot(kind='bar')
plt.title("Total VAT by City")
plt.ylabel("VAT Amount")
plt.show()

vat_city
```

```python
vat_customer_type = data.groupby('Customer type')['Tax 5%'].sum().sort_values()

plt.figure(figsize=(7,4))
vat_customer_type.plot(kind='bar')
plt.title("VAT Paid by Customer Type")
plt.ylabel("VAT Amount")
plt.show()

vat_customer_type
```
